```python
import requests
from bs4 import BeautifulSoup
import os
import time

BASE_URL = "https://www.nrb.org.np/category/monthly-statistics/page/{}/"
DOWNLOAD_DIR = r"C:\Users\aaros\OneDrive\Documents\Remittance_Analysis\data\nrb_raw"
os.makedirs(DOWNLOAD_DIR, exist_ok=True)

all_xlsx_urls = []

for page in range(1, 19):
    url = BASE_URL.format(page)
    r = requests.get(url, timeout=30)
    soup = BeautifulSoup(r.text, "html.parser")
    
    for a in soup.find_all("a", href=True):
        href = a["href"]
        if href.endswith(".xlsx"):
            all_xlsx_urls.append(href)
    
    time.sleep(1)

print(f"Found {len(all_xlsx_urls)} xlsx files")

for url in all_xlsx_urls:
    fname = url.split("/")[-1]
    fpath = os.path.join(DOWNLOAD_DIR, fname)
    if os.path.exists(fpath):
        continue
    try:
        r = requests.get(url, timeout=60)
        with open(fpath, "wb") as f:
            f.write(r.content)
        print(f"Downloaded: {fname}")
        time.sleep(0.5)
    except Exception as e:
        print(f"FAILED: {url} — {e}")
```

```python
import re

years = set()
for f in os.listdir(DOWNLOAD_DIR):
    match = re.search(r'20\d\d', f)  # finds 4-digit years like 2078, 2081
    if match:
        years.add(match.group())

print("BS years covered:", sorted(years))
```

```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook

# Adjust this to your actual downloads folder
DATA_DIR = r"C:\Users\aaros\Downloads"
OUTPUT_DIR = r"C:\Users\aaros\OneDrive\Documents\Remittance_Analysis\output"

import os
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
wb = load_workbook(f"{DATA_DIR}\\Balance-of-Payments-BPM5.xlsx", read_only=True)
ws = wb['BOP BPM5']
rows = list(ws.iter_rows(values_only=True))

# Row 3 (index 2) has fiscal years, Row 4 (index 3) has months
fy_row   = rows[2]   # e.g. '2000/01', None, None, ..., '2001/02', ...
mon_row  = rows[3]   # e.g. None, 'Aug', 'Sep', ..., None, 'Aug', ...
rem_row  = rows[31]  # Row 32 = Workers' remittances (0-indexed = 31)

# Build date → value mapping
# Fiscal years start at col 5 (0-indexed), each year = 12 columns
FISCAL_YEAR_MAP = {
    'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12,
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5,
    'Jun': 6, 'Jul': 7
}

records = []
fy_current = None

for col_idx in range(5, len(fy_row)):
    # Forward-fill fiscal year
    if fy_row[col_idx] is not None:
        fy_current = str(fy_row[col_idx]).strip()  # e.g. '2000/01'
    
    month_label = mon_row[col_idx]
    if month_label is None or fy_current is None:
        continue
    month_label = month_label.strip()
    if month_label not in FISCAL_YEAR_MAP:
        continue
    
    value = rem_row[col_idx]
    if value is None or value == '-':
        continue
    try:
        value = float(value)
    except (TypeError, ValueError):
        continue
    
    # Convert to Gregorian date
    # Nepali fiscal year 2000/01 starts mid-July 2000
    # Months Aug-Dec belong to the first calendar year in "YYYY/YY"
    # Months Jan-Jul belong to the second calendar year
    start_year = int(fy_current[:4])
    month_num = FISCAL_YEAR_MAP[month_label]
    cal_year = start_year if month_num >= 8 else start_year + 1
    
    date = pd.Timestamp(year=cal_year, month=month_num, day=1)
    records.append({'date': date, 'remittance': value})

wb.close()

df_remittance = pd.DataFrame(records).drop_duplicates('date').sort_values('date').reset_index(drop=True)
print(f"Remittance rows: {len(df_remittance)}")
print(f"Date range: {df_remittance['date'].min()} to {df_remittance['date'].max()}")
print(df_remittance.head())
```

```python
wb = load_workbook(f"{DATA_DIR}\\Exchange-rate.xlsx", read_only=True)
ws = wb['Time series']
rows = list(ws.iter_rows(values_only=True))

MONTH_MAP = {
    'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12,
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5,
    'June': 6, 'July': 7
}

records_exr = []
fy_current = None

for row in rows[4:]:  # skip 4 header rows
    if row[0] is not None:
        fy_current = str(row[0]).strip()
    
    month_raw = row[1]
    avg_mid   = row[4]  # Monthly Average Middle Rate column
    
    if month_raw is None or fy_current is None or avg_mid in (None, '-', ''):
        continue
    
    month_label = month_raw.strip()
    if month_label not in MONTH_MAP:
        continue
    
    try:
        avg_mid = float(avg_mid)
    except (TypeError, ValueError):
        continue
    
    start_year = int(fy_current[:4])
    month_num = MONTH_MAP[month_label]
    cal_year = start_year if month_num >= 8 else start_year + 1
    
    date = pd.Timestamp(year=cal_year, month=month_num, day=1)
    records_exr.append({'date': date, 'exchange_rate': avg_mid})

wb.close()

df_exr = pd.DataFrame(records_exr).drop_duplicates('date').sort_values('date').reset_index(drop=True)
# Filter to your study period
df_exr = df_exr[df_exr['date'] >= '2000-01-01']
print(f"Exchange rate rows: {len(df_exr)}")
print(f"Date range: {df_exr['date'].min()} to {df_exr['date'].max()}")
print(df_exr.head())
```

```python
wb = load_workbook(f"{DATA_DIR}\\MIgrant-Workers_.xlsx", read_only=True)
ws = wb['Migrant Worker']
rows = list(ws.iter_rows(values_only=True))

records_mw = []
for row in rows[2:]:  # skip 2 header rows
    date_val = row[0]
    total    = row[3]  # "Total Worker's Outflow" — monthly, not cumulative
    
    if not isinstance(date_val, pd.Timestamp) and not hasattr(date_val, 'year'):
        continue
    if total is None or isinstance(total, str):
        continue
    
    try:
        total = float(total)
    except (TypeError, ValueError):
        continue
    
    date = pd.Timestamp(date_val).replace(day=1)
    records_mw.append({'date': date, 'dofe_departures': total})

wb.close()

df_mw = pd.DataFrame(records_mw).drop_duplicates('date').sort_values('date').reset_index(drop=True)
print(f"Migrant worker rows: {len(df_mw)}")
print(f"Date range: {df_mw['date'].min()} to {df_mw['date'].max()}")
print(df_mw.head())
```
